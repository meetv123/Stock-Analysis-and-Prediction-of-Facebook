```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data from excel file
data = pd.read_excel('/Users/meetvanani/Downloads/MetaDataSet.xlsx')

print(data)

# Data cleaning and preprocessing
data.dropna(inplace=True) # Drop missing values
data.drop_duplicates(inplace=True) # Drop duplicates
data['Date'] = pd.to_datetime(data['Date']) # Convert date column to datetime format

# Basic descriptive statistics
stats = data[['Bid', 'Ask', 'Volume']].describe()
```

```python
# Visualization
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Bid'], label='Bid Price')
plt.plot(data['Date'], data['Ask'], label='Ask Price')
plt.title('Facebook(Meta) Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(data['Date'], data['Volume'])
plt.title('Facebook(Meta) Trading Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(data['Bid'], data['Volume'])
plt.title('Bid Price vs. Trading Volume')
plt.xlabel('Bid Price')
plt.ylabel('Volume')
plt.show()
```

```python
# Statistical analysis
corr = data[['Bid', 'Volume']].corr()

# Predictive modeling
X = data[['Volume']] # Independent variable
y = data['Bid'] # Dependent variable
regressor = LinearRegression()
regressor.fit(X, y) # Fit the model
y_pred = regressor.predict(X) # Predict the values
plt.figure(figsize=(8, 8))
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.title('Bid Price vs. Trading Volume (Linear Regression)')
plt.xlabel('Volume')
plt.ylabel('Bid Price')
plt.show()
```

```python
# Communicating findings
print(stats)
print(corr)
```
